```python
import json
```

```python
filename = "random_sentence_16.json"
processing_filename = "random_sentence_16_processing.json"

encoded = {}
with open(filename, "r") as file:
    encoded = json.load(file)
```

```python
emfreq_hex_map = {
    121.57: "6467006a",
    102.57: "64640066",
    97.254: "64660068",
    94.974: "64650067",
    93.780: "64650066",
    93.075: "64610061",
    656.279: "648300b3" ,
    486.135: "647f009d",
    434.047: "647c0095",
    410.173: "647c0093",
    397.008: "647b0091",
    1875.: "640800ff",
    1282.: "646a00ff",
    1094.: "647900e6",
    1005.: "647c00de",
    4051.: "6400ff17",
    2625.: "640089ff",
    2166.: "64003bff",
    7460.: "64ff7700",
    4654.: "645aff00",
    12370.: "64610000"
}
```

```python
data = []

for j in range(len(encoded)):
    
    word = []
    for i in range(6):
        m_key = "m" + str(i+1)
        m = encoded[j][m_key]

        emfreq = m["phoneme"]["emission"]["emfreq"]

        hex_code = emfreq_hex_map[emfreq]
        hap1 = m["phoneme"]["haptics"]["outer"]
        hap2 = m["phoneme"]["haptics"]["inner"]

        sounds = len(m["phoneme"]["sounds"])

        m_dict = {
            "color": hex_code,
            "hap1": hap1,
            "hap2": hap2,
            "numsounds": sounds
        }

        word.append(m_dict)

    data.append(word)
    
p = {"data": data}
with open(processing_filename, "w") as file:
    json.dump(p, file, indent = 4)
```
