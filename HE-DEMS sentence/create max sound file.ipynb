```python
import maxpy as mp
import json
```

```python
delay_length = 100
```

```python
with open("random_sentence.json", "r") as file:
    sentence_data = json.load(file)
```

```python
sentence_data[0]['m1']['soundshape']
```

```python
sound_data = []

morphs = ["m1", "m2", "m3", "m4", "m5", "m6"]

for word in sentence_data:
    
    word_data = []
    for morph in morphs:
        len_sounds = len(word[morph]['phoneme']['sounds'])
        morph_data = []
        for i in range(len_sounds):
            soundfreq = word[morph]['phoneme']['sounds'][i]['soundfreq']
            shape = word[morph]['soundshape'][i]

            morph_data.append([soundfreq, shape])
        
        word_data.append(morph_data)
    
    sound_data.append(word_data)
```

```python
#index sound data by [word_num][morphh_num], both starting at 0 
#list of lists [freq, shape ]
```

```python
sound_data[0][1]
```

```python
#make morph patches 

for i in range(len(sound_data)):
    for j in range(6):
        morph_sounds = sound_data[i][j]
        num_sounds = len(morph_sounds)
        
        morph_patch = mp.MaxPatch()
        inlet, outlet = morph_patch.place("inlet", "outlet")
        
        prev_delay = None
        for k in range(num_sounds):
        
            #sound info 
            s_type = morph_sounds[k][1]
            if(s_type) == 'free':
                s_type = 'spike' #lore moment: replace "FREE" w/ "SPIKE" in billboards 
            s_res_string = "reson~ 25. " + str(morph_sounds[k][0]) + " 230."

            #placing shape & reson 
            s = morph_patch.place(s_type)[0]
            s_res = morph_patch.place(s_res_string)[0]
            
            #place delays, if necessary
            if k > 0:
                delay = morph_patch.place("delay " + str(delay_length))[0]
        
            #make connections 
            morph_patch.connect(( s.outs[0], s_res.ins[0] ),
                                (s_res.outs[0], outlet.ins[0] ))
            
            if k == 0:
                morph_patch.connect( (inlet.outs[0], s.ins[0] ))
            elif k == 1:
                morph_patch.connect( (inlet.outs[0], delay.ins[0] ),
                                     (delay.outs[0], s.ins[0] ))
                prev_delay = delay
            elif k == 2:
                morph_patch.connect( (prev_delay.outs[0], delay.ins[0] ),
                                     (delay.outs[0], s.ins[0] ))
            
        morph_patch.save("word" + str(i+1) + "_morph" + str(j+1))
        
        
```
