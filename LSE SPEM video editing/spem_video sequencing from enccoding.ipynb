```python
#load file from json
import json
with open("sentence-encoding.json", "r") as file:
    sentence_encoding = json.load(file)
```

```python
def get_events_each_half_count(info_list):
    sum = 0
    events = []
    i = 0
    
    while sum < 123:
        event = info_list[i]
        sum += event[1] 
        
        if i == 0:
            for j in range(int(event[1]*2)):
                events.append(event[0])
        
        else:
            for j in range(len(events), int(sum*2)):
                events.append(event[0])
                
        i += 1
        
    return events
```

```python
all_events = {
}

for i in range(2, 8):
    
    hand_events = get_events_each_half_count(sentence_encoding[str(i)]["hand"])
    slap_events = get_events_each_half_count(sentence_encoding[str(i)]["slap"])
    eyes_events = get_events_each_half_count(sentence_encoding[str(i)]["eyes"])
    voice_events = get_events_each_half_count(sentence_encoding[str(i)]["voice"])
    
    all_events[i] = {
        "hand":hand_events,
        "slap":slap_events,
        "eyes":eyes_events,
        "voice":voice_events
    }
    
    
```

```python
all_info_rows = []

for i in range(len(all_events[2]['hand'])):
    info = []
    for j in range(2, 8):
        node = all_events[j]
        info.append(node['hand'][i])
        info.append(node['slap'][i])
        info.append(node['eyes'][i])
        info.append(node['voice'][i])
        info.append(" ")
    all_info_rows.append(info)
```

```python
import csv
with open('test.csv', "w") as file:
    writer = csv.writer(file)
    for row in all_info_rows:
        writer.writerow(row)
```
